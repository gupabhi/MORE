```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
input = 'data/'
df = pd.read_excel(input + 'structured-dataset.xlsx', header=1, usecols='A,I,L', sheet_name='Mos_OR')
```

```python
df.info()
```

```python
response_df = df.groupby(['odor', 'or'], as_index=False).response.mean()
```

```python
odor = pd.read_csv(input + 'odor-selected-descriptors-data.csv', sep='#')
```

```python
merged_df = response_df.merge(odor, how='inner', left_on='odor', right_on='odor_sheet_name')
```

```python
merged_df
```

```python
data=merged_df.drop(['odor_sheet_name', 'odor_pubchem_name', 'cid', 'iupac_name', 'isomeric_smiles'], axis=1)
```

```python
data.nunique()
```

```python
from sklearn.preprocessing import StandardScaler
data[data.columns[2:]] = StandardScaler().fit_transform(data[data.columns[2:]])
```
